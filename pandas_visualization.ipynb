# Pandas Visualization

```python
import pandas as pd

iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",  names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

#print(iris)

print(iris.head())
```

## Scatter Plot

```python
iris.plot.scatter(x='sepal_length', y='sepal_width', title='Iris Dataset')
```

```python
iris.drop(['class'], axis=1).plot.line(title='Iris Dataset')
```

```python
iris.describe()
```

## Histogram

```python
iris.plot.hist()
```

### Its also really easy to create multiple histograms.

```python

iris.plot.hist(subplots=True, layout=(4,1), figsize=(5, 10), bins=20)
```

## Bar Chart

```python
iris['class'].value_counts().sort_index().plot.bar()
```

```python
iris.groupby(['class']).count()
```

```python
iris['class'].value_counts().sort_index().plot.barh()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris, hue='class')
```

```python
sns.distplot(iris['sepal_length'],bins=10,kde=True)
```

```python
sns.countplot(iris['class'])
```

```python
sns.boxplot('sepal_length',data=iris)
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr(),annot=True)
```
